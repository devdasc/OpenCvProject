```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread(r'C:\Users\devda\PythonProjects\OpenCv\OpenCvProject\images\nature.jpg')
cv2.imshow('gray image', image)
cv2.waitKey()
histogram = cv2.calcHist([image],[0],None,[256],[0,255])
plt.figure()
plt.xlim([0,255])
plt.xlabel('pixel intensities')
plt.ylabel('no of pixels')
plt.plot(histogram,color='r')
plt.show()
```
